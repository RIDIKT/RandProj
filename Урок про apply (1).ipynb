```python
# Премия 5% от Salary
## PerformanceScore  < 4 тогда получает 50%
```

```python
def calc_bonus(row):
    base = 0.05 * row['Salary']
    #bonus = base if row['PerformanceScore'] >= 4 else base * 0.5
    if row['PerformanceScore'] >= 4:
        return base
    else:
        return base * 0.5
```

```python
df['CalculatedBonus'] = df.apply(calc_bonus, axis=1)
```

```python
df
```

```python
# Категоризация сотрудников:
## Топовый сотрудник: зарплата > 75000 и оценка >= 4
## Требует улучшений: зарплата <= 75000 и оценка ниже 4
## Средний сотрудник
```

```python
def categorize_employee(row):
    if row['Salary'] > 75_000 and row['PerformanceScore'] >= 4:
        return 'Top'
    elif row['Salary'] <= 75_000 and row['PerformanceScore'] < 4:
        return 'Needs Improvement'
    else:
        return 'Average'
```

```python
df['Category'] = df.apply(categorize_employee, axis=1)
```

```python
df
```

```python
def conditional_bonus(row, bonus_rate, threshold):
    if row['Salary'] > threshold:
        return row['Salary'] * bonus_rate
    return 0
```

```python
data = {
    'EmployeeID': [101, 102, 103, 104, 105],
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Ethan'],
    'Department': ['Sales', 'Engineering', 'HR', 'Engineering', 'Sales'],
    'Salary': [55000, 80000, 50000, 90000, 60000],
    'JoinDate': ['2020-01-15', '2019-07-23', '2021-03-01', '2018-12-12', '2020-06-30'],
    'PerformanceScore': [4.5, 3.8, 4.2, 4.0, 3.9]
}
df = pd.DataFrame(data)
df['JoinDate'] = pd.to_datetime(df['JoinDate'])
```

```python
df
```

```python
# Форматирование имени сотрудника с префиксом
```

```python
df['FormalName'] = df['Name'].apply(lambda x: f"Employee: {x}")
df[['Name', 'FormalName']]
```

```python
# Вычисление "стажа" в годах на основе JoinDate
```

```python
current_year = datetime.now().year
df['Tenure_Years'] = df['JoinDate'].apply(lambda d: current_year - d.year)
df[['JoinDate', 'Tenure_Years']]
```

```python
# Извлечение года из даты приема
```

```python
df['Year'] = df['JoinDate'].apply(lambda d: d.year)
df[['Name', 'JoinDate', 'Year']]
```

```python
# Вычисление количества дней стажа
```

```python
df['Tenure_Days'] = df['JoinDate'].apply(lambda d: (datetime.now() - d).days)
df[['Name', 'JoinDate', 'Tenure_Days']]
```

```python
# Проверка, пришёл ли сотрудник до 2020 года
```

```python
df['JoinedBefore2020'] = df['JoinDate'].apply(lambda d: d.year < 2020)
df[['Name', 'JoinDate', 'JoinedBefore2020']]
```

```python
# Форматирование даты в строку формата "DD-MM-YYYY"
```

```python
df['JoinDate_Str'] = df['JoinDate'].apply(lambda d: d.strftime('%d-%m-%Y'))
df[['Name', 'JoinDate', 'JoinDate_Str']]
```
